### II. Retour sur les boucles
A la fin de la séance 1, nous avons travaillé sur la boucle "for element in liste". Nous allons approfondir les boucles cette semaine pour permettre plus de possiblités du codage. 

#### Boucles for in range()
>Appeler et exécuter la fonction ci-dessous :

```python
def affiche_suite() : 
    n = int(input("Saisir un entier positif : "))
    for i in range(n) : 
        print(i)
```

***Vocabulaire*** : le mot clé ***range*** signifie ***intervalle***. 

Dans cette boucle, ***i*** varie dans l'intervalle \[0, 5\[. La forme intégrale de ***range()*** est ***range(début, fin, pas)***. 

***Attention*** : la valeur de fin est exclue des itérations, c'est-à-dire que la boucle va jusqu'à ***i=fin-1***. Les valeurs de *début* et de *pas* ne sont pas toujours précisées. Dans le cas échéant, les valeurs par défaut sont appliquées : 0 pour *début* et 1 pour *pas*.

La boucle ***while*** est souvent utilisée pour le contrôle de saisie. Au début du TP, on a demandé à l'utilisateur de saisir un entier positif. Il est donc tout à fait possible vérifier si la saisie est conforme. 
> Exécutez la fonction suivante :

```python
def affiche_suite() : 
    while True : 
        n = int(input("Saisir un entier positif : "))
        if n > 0 :
            break
    for i in range(n) : 
        print(i)
```

La boucle ***while True*** est souvent utilisée dans les mini-jeux codés. La boucle continue à exécuter jusqu'à tomber sur (une des) conditions d'arrêt. Dans cet exemple, la boucle continue à demande une saisie jusqu'à ce que ***n > 0***.
