####  2. Deuxième méthode : la méthode .format()

Chaque **paire d'accolades** désigne **un champ** qui est rempli avec **la valeur d'un des paramètres** de la méthode format().

Exemple:
> Exécutez les cellules suivantes :


```python
print("Je m'appelle {} et j'ai {} ans".format("Bob",10))
```

```python
print("Je m'appelle {0} et j'ai {1} ans".format("Bob",10))
```

### II. Découverte  des dictionnaires

Un dictionnaire en python est une sorte de liste mais au lieu d'utiliser des index , on utilise des clés alphanumériques. 

#### 1. Création d'un dictionnaire vide
avec des accolades, ou le mot clé **dict**.
> Affichez les types et tailles de fruit1 et fruit2 dans le cellule suivante:


```python
fruit1={}
fruit2=dict()
```

#### 2. Création d'un dictionnaire  
En rajoutant des clés, et en associant à ces clés, des valeurs.

> Afficher les dictionnaires fruit1 et fruit2 dans la cellule suivante et créer un autre dictionnaire fruit3 décrivant une banane en reprenant les mêmes clés.

```python
fruit1['nom'] = 'pomme'
fruit1['couleur'] = 'vert'
fruit1['prix'] = 1.2
#ou
fruit2={'nom':'orange','couleur':'orange','prix':3.4}
```

#### 3. Liste des dictionnaires

Il est tout à fait possible de créer une liste de dictionnaires, par exemple :

```python
liste_fruits = [fruit1, fruit2]

print(liste_fruits)
```

Les **clés** du dictionnaire fruit1 sont:
    - 'nom'
    - 'couleur'
    - 'prix'
A chaque clé correspond une **valeur**:
    - 'nom' => 'pomme'
    - 'couleur' => 'vert'
    - 'prix' => 1.2

#### 4. Accès aux valeurs et modifications

Pour récupérer la valeur d'une clé donnée et/ou la modifier, il suffit d'utiliser une syntaxe du style dictionnaire[cle].  
Exemple:  

```python
print(fruit1['nom'])
print(liste_fruits[1]['prix'])
```

```python
del liste_fruits[0]['couleur']

print(fruit1)
```

#### 5. Afficher les clés et les valeurs

Grâce à la méthode ***.items()***, il est également possible d'afficher les clés (keys) et les valeurs (values) d'un dictionnaire. Exécutez et comparez les deux boucles suivantes : 

```python
for item in fruit1.items() : 
    print(item)
    
for cle, valeur in liste_fruits[1].items() : 
    print(cle, valeur)
```

Dans la première boucle, les *item* affichés sont du typle ***tuple*** et contiennent deux informations : les clés et les valeurs. Ces informations peuvent également être affichés séparément, comme dans la deuxième boucle. 

Il existe également des méthodes qui permettent de récupérer uniquement les clés (***.keys()***) ou les valeurs (***.values()***). Par exemple : 

```python
print(fruit1.keys())
for cle in fruit1.keys() : 
    print(cle)
```

### III. Banque aux astuces

1. Utilisation de l'opérateur d'appartenance : **in , not in**

L'instruction ***a in B*** retourne un booléen pour indiquer si ***a*** appartient à ***B***. Cette astuce peut être utilisés dans plusieurs circonstances :

- Tester si un élément est dans une liste

```python
liste_1 = ["bonjour", 1.5]
print(2 in liste_1)
```

- Tester si une chaîne de caractères et dans une autre chaîne

```python
str1 = "Bonjour, nous sommes en 2023."
str2 = "sommes"
print(str2 in str1)
print(str2 not in str1)
```

On peut retrouver ces deux opérateurs d'appartenances dans **les tests et dans les boucles bien sûr!**

```python
if (5 not in liste_1):
    print("5 n'est pas dans la liste")
    
for i in ["a","b","c","d"]:
    print(i)
```
