#### Les arguments (entrées, sorties)

Voici une fonction très simple.

```python
def FonctionSimple(n):
    return 2*n
```

Vous l'avez compris, la fonction prend en entrée un nombre n et renvoie la valeur du double de ce nombre. C'est un peu comme définir la fonction f(x) = 2x en mathématiques.

Lorsque l'on lance le code précédent, rien ne se passe. C'est normal, vous avez défini une fonction, mais vous n'avez pas demandé d'afficher ou de calculer quoi que ce soit. Vous avez simplement signalé "Si j'appelle la fonction FonctionSimple sur un entier, je veux obtenir la valeur du double de cet entier." La fonction, une fois définie, peut être utilisée ultérieurement.

```python
nombre = 76
nombre2 = FonctionSimple(nombre)
print(nombre2)
```

### Tests : if , elif, else (si , sinon si, sinon)

> Exécutez la cellule suivante et en modifiant les valeurs de a et b, complétez le code et vérifiez le comportement de la structure ***if elif else***  

```python
def fonctionTest() :
    a, b = 2, 3 #affectation simultannée des variables a et b
    if a > b:
        return str(a)+" ... "+str(b)
    elif a == b:
        return str(a)+" ... "+str(b)
    else:
        return str(a)+" ... "+str(b)
        
print(fonctionTest())
```

### Boucles for (pour)
> Observez le fonctionnement de la première fonction suivante. Puis, complétez la deuxième fonctions pour afficher chaque éléments de la liste ***ferme***. Faites en sorte que chaque ligne corresponde à un enclos et les informations du même enclos soient séparé par des espaces.

```python
# Testez le code suivant
def afficherListe(lst) :
    for element in lst:
        print(element)
    
liste = [10, 20, 30]
afficherListe(liste)
```

```python
# Complétez le code suivant et lancez un test !
def afficherListe2D(lst) : 
    for ligne in lst : 
        enclos = "" # Réinitaliser le texte à afficher à vide
        for element in ligne : 
            enclos = enclos + " " + str(element)
        print(enclos)
            
ferme = [[1,'cochon(s)', 4],
         [2,'vache(s)', 21],
         [3,'mouton(s)', 10],
         [4,'poule(s)', 25]]

afficherListe2D(ferme)
```

**4. Recherche d'un nombre entre 1 et 100 par dichotomie +++**   
La recherche par dichotomie est une méthode qui consiste à diviser (en général en parties égales) un problème pour en trouver la solution.
Pour arriver rapidement à deviner le nombre, l'astuce consiste à prendre à chaque fois la moitié de l'intervalle dans lequel se trouve le nombre. 
Voici le détail des différentes étapes :  
le nombre se trouve entre 1 et 100, on propose 50 (100 / 2).  
le nombre se trouve entre 50 et 100, on propose 75 ( 50 + (100-50)/2 ).  
le nombre se trouve entre 75 et 100, on propose 87 ( 75 + (100-75)/2 ).  
le nombre se trouve entre 75 et 87, on propose 81 ( 75 + (87-75)/2 ).  
le nombre se trouve entre 75 et 81, on propose 78 ( 75 + (81-75)/2 ).  
le nombre se trouve entre 78 et 81, on propose 79 ( 78 + (81-78)/2 ).
